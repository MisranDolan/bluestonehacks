```python
#Cleaning the strings on the street names
#Parsing through the names and removing unnecessary words

s = "Original Route Alias: "
l = len(s)
speed = speed.dropna(axis = 0, subset=["COMMENT_TE"])

for i in speed.index:
    #Should check all unique values, because there's a few cells that don't start with Original Route Alias
    speed.at[i,"COMMENT_TE"] = speed.at[i,"COMMENT_TE"][l:]
    if (speed.at[i,"COMMENT_TE"].find("In accordance") != -1):
        find = speed.at[i,"COMMENT_TE"].find("In accordance")
        speed.at[i,"COMMENT_TE"] = speed.at[i,"COMMENT_TE"][:find]
        speed.at[i,"COMMENT_TE"] = speed.at[i,"COMMENT_TE"].replace('\r\n','')
```
